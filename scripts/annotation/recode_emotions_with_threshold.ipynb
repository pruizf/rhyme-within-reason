```python
import pandas as pd
import numpy as np
```

```python
dfname = "../../data/dataframe-all_lem_sets_emos_with_nrc_filt.tsv"
```

```python
emonames = sorted({"anger", "anticipation", "disgust", "fear", "joy", "sadness",
                   "surprise", "trust"})
```

```python
colnames_call = [f"{x}_call" for x in emonames]
colnames_echo = [f"{x}_echo" for x in emonames]
```

```python
threshold = 0.5

for cn in colnames_call + colnames_echo:
    df_echo_full[cn] = np.where(df_echo_full[cn] >= threshold, df_echo_full[cn], np.nan)
```

```python
df_echo_full
```

```python
out_fname = f"../../data/emotion_combination_{str(threshold).replace('.','')}.tsv"
```

```python
df_echo_full.to_csv(out_fname, index=False, sep="\t")
```
